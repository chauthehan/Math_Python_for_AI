## What is a While Loop?

A **while loop** repeats a block of code as long as a condition is `True`. Think of it like this:

- **If statement**: "IF the condition is true, do something ONCE"
- **While loop**: "WHILE the condition is true, KEEP DOING something"

Here's the basic syntax:

```python
while condition:
    # code to repeat
    # this keeps running as long as condition is True
```


### Simple Example: Counting Down

Let's start with a simple countdown from 5 to 1:


````python
count = 5

while count > 0:
    print(f"Count: {count}")
    count = count - 1  # Decrease count by 1
    
print("Blast off! ðŸš€")
````

**What happened here?**

1. We start with `count = 5`
2. The loop checks: "Is count > 0?" â†’ Yes (5 > 0), so it runs the code
3. It prints "Count: 5" and decreases count to 4
4. The loop checks again: "Is count > 0?" â†’ Yes (4 > 0), so it runs again
5. This continues until count becomes 0
6. When count = 0, the condition "count > 0" is False, so the loop stops
7. Finally, it prints "Blast off!"

**Important:** The value of `count` must change inside the loop, otherwise it would run forever!


### Another Example: Accumulating Points

Let's say you're playing a game and want to keep adding points until you reach 100:


````python
score = 0

while score < 100:
    score = score + 15
    print(f"Current score: {score}")
    
print(f"Congratulations! You reached {score} points!")
````

## While Loops for Chatbots: The while True Pattern

For chatbots, we usually want them to keep responding until the user decides to quit. The most common pattern is:

```python
while True:
    # get user input
    # respond
    # check if user wants to quit
    # if yes, break out of the loop
```

The condition `True` is always true, so the loop would run forever... **unless** we use the `break` statement to exit!


### Simple Chatbot with While Loop


````python
print("Bot: Hello! I'm your friendly chatbot. Type 'bye' to exit.")

while True:
    user_message = input("You: ")
    
    # Check if user wants to quit
    if user_message.lower() == "bye":
        print("Bot: Goodbye! Have a great day!")
        break  # Exit the loop
    
    # Otherwise, respond
    print(f"Bot: You said: '{user_message}'")
````

**What's happening here?**

1. The loop starts with `while True:` - this would run forever
2. Inside the loop, we get user input
3. We check if the user typed "bye" (using `.lower()` to handle "Bye", "BYE", etc.)
4. If they typed "bye", we print a goodbye message and use `break` to exit the loop
5. If they didn't type "bye", we respond and the loop continues

Try running the cell above! Type different messages and watch the chatbot respond. When you're ready to stop, type "bye".


### The break Statement

The `break` statement is used to **exit a loop immediately**. When Python sees `break`, it stops the loop and continues with the code after the loop.

```python
while True:
    # some code
    if some_condition:
        break  # Exit the loop right now!
    # more code
```

This is perfect for chatbots because you want them to keep running until the user says to stop!

## âš ï¸ Warning: Infinite Loops!

An **infinite loop** is a loop that never stops. This happens when the condition is always `True` and there's no `break` statement.

**Bad Example** (Don't run this!):
```python
count = 5
while count > 0:
    print(count)
    # Forgot to decrease count!
    # This will print 5 forever!
```

If you accidentally create an infinite loop:
- Press the **Stop** button in Jupyter
- Or use **Ctrl + C** in the terminal

**How to avoid infinite loops:**
1. Make sure your condition will eventually become `False`
2. If using `while True:`, always include a `break` statement
3. Double-check that variables in your condition are being updated


## Advanced Chatbot: Personality Quiz

Here's a more complex example that combines everything we've learned - a personality quiz chatbot!


````python
print("Bot: Welcome to the Personality Quiz Chatbot!")
print("Bot: Answer with 'a' or 'b' for each question.")
print()

# Ask for name
print("Bot: First, what's your name?")
user_name = input("You: ").strip().title()
print(f"Bot: Nice to meet you, {user_name}!")
print()

score = 0
question_number = 1
max_questions = 3

while question_number <= max_questions:
    print(f"--- Question {question_number} of {max_questions} ---")
    
    if question_number == 1:
        print("Bot: Do you prefer:")
        print("  a) Spending time outdoors")
        print("  b) Staying inside with a book")
        answer = input("You: ").lower().strip()
        if answer == "a":
            score = score + 1
    
    elif question_number == 2:
        print("Bot: At a party, do you:")
        print("  a) Talk to lots of people")
        print("  b) Stick with a few close friends")
        answer = input("You: ").lower().strip()
        if answer == "a":
            score = score + 1
    
    elif question_number == 3:
        print("Bot: Would you rather:")
        print("  a) Try something new and exciting")
        print("  b) Do something you already enjoy")
        answer = input("You: ").lower().strip()
        if answer == "a":
            score = score + 1
    
    question_number = question_number + 1
    print()

# Calculate result
print(f"Bot: Thanks for completing the quiz, {user_name}!")
print(f"Bot: Your score: {score} out of 3")

if score >= 2:
    print("Bot: You're an ADVENTURER! You love excitement and new experiences!")
else:
    print("Bot: You're a HOMEBODY! You appreciate comfort and familiar things!")
    
print("\nBot: Thanks for chatting with me!")
````

This quiz chatbot:
- Asks for the user's name and formats it nicely
- Uses a while loop to go through 3 questions
- Keeps track of the user's score
- Uses `question_number` as a counter to track progress
- Gives a personality result at the end
